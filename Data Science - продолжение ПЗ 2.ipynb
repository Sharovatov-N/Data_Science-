## Практическое задание к уроку №2.

### 1. В чём различие между зависимыми и независимыми выборками?

Зависимые и независимые выборки. (*на гуглил, но согласен с этим*) 

Обычна ситуация исследования, когда интересующее исследователя свойство изучается на двух или более выборках с целью их дальнейшего сравнения. Эти выборки могут находиться в различных соотношениях — в зависимости от процедуры их организации. Независимые выборка характеризуются тем, что вероятность отбора любого испытуемого одной выборки не зависит от отбора любого из испытуемых другой выборки. Напротив, зависимые выборки характеризуются тем, что каждому испытуемому одной выборки поставлен в соответствие по определенному критерию испытуемый из другой выборки.

ПРИМЕР

Наиболее типичный пример зависимых выборок — повторное измерение свойства (свойств) на одной и той же выборке после воздействия (ситуация до-после). В этом случае выборки (одна — до, другая — после воздействия) зависимы в максимально возможной степени, так как они включают одних и тех же испытуемых. Могут быть и более слабые варианты зависимости. Например, мужья — одна выборка, их жены — другая выборка (при исследовании, например, их предпочтений). Или дети 5-7 лет — одна выборка, а их братья или сестры-близнецы — другая выборка.

### 2. Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

Если выборка непрерывных данных имеет нормальное распределение, то для такой выборке применяют параметрические статистические критерии.

Если выборка бинарная, категориальная или непрерывных данных но ненормально распределенных, то для таких выборок проименяют непараметрические аналоги статистических критериев.
